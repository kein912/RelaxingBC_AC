```python
from numpy import genfromtxt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import math
```

```python
def cal_relax_level(df, dataClass):
    # 分離 'pre' 和 'post' 數據
    df_pre = df[df.stateLabel == 0]
    df_post = df[df.stateLabel == 2]

    # 確保兩個 DataFrame 的行順序相同，以便於計算
    df_pre = df_pre.sort_values(by='data_name').reset_index(drop=True)
    df_post = df_post.sort_values(by='data_name').reset_index(drop=True)

    # 選擇需要計算的列
    columns_to_calculate = df.columns[1:-2]  # 假設最後兩列是 'addictLabel' 和 'stateLabel'

    # 計算放鬆程度
    # relaxation_levels = (df_post[columns_to_calculate] - df_pre[columns_to_calculate]) / df_pre[columns_to_calculate]
    relaxation_levels = (df_post[columns_to_calculate] - df_pre[columns_to_calculate])

    # 將結果加入到一個新的 DataFrame
    relaxation_df = pd.DataFrame(relaxation_levels, columns=columns_to_calculate)
    relaxation_df['data_name'] = df_pre['data_name'].str.replace('_Pre', '')  # 添加數據名稱

    # 调整列的顺序，使 'data_name' 位于第一列
    column_order = ['data_name'] + list(columns_to_calculate)
    relaxation_df = relaxation_df[column_order]

    print(relaxation_df)
    relaxation_df.to_excel('Feature3group/' + dataClass + '_relaxation_level_sub.xlsx', index=False)
```

```python
eeg_data = pd.read_csv("Feature/EEG_PCR.csv", header=0, index_col=0)
# eeg_data3group = pd.read_csv("Feature3group/EEG_PCR.csv", header=0, index_col=0)
df = pd.DataFrame(eeg_data)
# df = pd.DataFrame(eeg_data3group)
# 使用concat函数上下合并DataFrame
# df = pd.concat([df1, df2], axis=0, ignore_index=True)

df = df[df['stateLabel'] != 3]

# 計算 theta/beta 的比值
df['FC3_theta/beta'] = df['FC3_theta'] / df['FC3_beta']
df['FCz_theta/beta'] = df['FCz_theta'] / df['FCz_beta']
df['Pz_theta/beta'] = df['Pz_theta'] / df['Pz_beta']
df['Oz_theta/beta'] = df['Oz_theta'] / df['Oz_beta']
eeg_data3group = df[['data_name', 'FC3_theta/beta', 'FCz_theta/beta', 'Pz_theta/beta', 'Oz_theta/beta', 'addictLabel', 'stateLabel']]

# 分離 'pre' 和 'post' 數據
df_pre = eeg_data3group[eeg_data3group.stateLabel == 0]
df_post = eeg_data3group[eeg_data3group.stateLabel == 2]

# 確保兩個 DataFrame 的行順序相同，以便於計算
df_pre = df_pre.sort_values(by='data_name').reset_index(drop=True)
df_post = df_post.sort_values(by='data_name').reset_index(drop=True)

# 選擇需要計算的列
columns_to_calculate = eeg_data3group.columns[1:-2]  # 假設最後兩列是 'addictLabel' 和 'stateLabel'

# 計算放鬆程度
# relaxation_levels = (df_post[columns_to_calculate] - df_pre[columns_to_calculate]) / df_pre[columns_to_calculate]
relaxation_levels = (df_post[columns_to_calculate] - df_pre[columns_to_calculate])

# 將結果加入到一個新的 DataFrame
relaxation_df = pd.DataFrame(relaxation_levels, columns=columns_to_calculate)
relaxation_df['data_name'] = df_pre['data_name'].str.replace('_Pre', '')  # 添加數據名稱

# 调整列的顺序，使 'data_name' 位于第一列
column_order = ['data_name'] + list(columns_to_calculate)
relaxation_df = relaxation_df[column_order]

print(relaxation_df)
relaxation_df.to_excel('Feature/' + 'EEG_PCR_Ratio_relaxation_level_sub.xlsx', index=False)
# relaxation_df.to_excel('Feature3group/' + 'EEG_PCR_Ratio_relaxation_level_sub.xlsx', index=False)
```

# AVG & STD

```python
def cal_relax_level_avgstd(df):
    data_A = df[df['data_name'].str.contains('A')].drop('data_name', axis=1)
    data_N = df[df['data_name'].str.contains('N')].drop('data_name', axis=1)

    mean_A = data_A.mean().round(3)
    # variance_A = data_A.var().round(3)
    std_A = data_A.std().round(3)

    mean_N = data_N.mean().round(3)
    # variance_N = data_N.var().round(3)
    std_N = data_N.std().round(3)

    # 合併平均值和方差成 "Mean (Var)" 形式
    # mean_var_A = mean_A.astype(str) + " (" + variance_A.astype(str) + ")"
    # mean_var_N = mean_N.astype(str) + " (" + variance_N.astype(str) + ")"
    mean_std_A = mean_A.astype(str) + " (" + std_A.astype(str) + ")"
    mean_std_N = mean_N.astype(str) + " (" + std_N.astype(str) + ")"
    mean_std = mean_A.astype(str) + " (" + std_A.astype(str) + ")" + "&" + mean_N.astype(str) + " (" + std_N.astype(str) + ")"

    # 將結果整理成DataFrame
    # results_A_mean_var = pd.DataFrame({'Mean_Var_A': mean_var_A})
    # results_N_mean_var = pd.DataFrame({'Mean_Var_N': mean_var_N})
    results_A_mean_std = pd.DataFrame({'Mean_Std_A': mean_std_A})
    results_N_mean_std = pd.DataFrame({'Mean_Std_N': mean_std_N})
    results_mean_std = pd.DataFrame({'Mean_Std': mean_std})

    # combined_results = pd.concat([results_A_mean_var, results_N_mean_var], axis=1)
    combined_results = pd.concat([results_A_mean_std, results_N_mean_std, results_mean_std], axis=1)

    display(combined_results)
```

```python
def cal_relax_level_avgstd_3group(type):
    data = pd.read_excel("Feature/%s_relaxation_level_sub.xlsx"%(type))
    data_A = data[data['data_name'].str.contains('A')].drop('data_name', axis=1)
    data3group = pd.read_excel("Feature3group/%s_relaxation_level_sub.xlsx"%(type))
    data_V = data3group[data3group['data_name'].str.contains('V')].drop('data_name', axis=1)
    data_C = data3group[data3group['data_name'].str.contains('C')].drop('data_name', axis=1)

    mean_A = data_A.mean().round(3)
    std_A = data_A.std().round(3)

    mean_V = data_V.mean().round(3)
    std_V = data_V.std().round(3)

    mean_C = data_C.mean().round(3)
    std_C = data_C.std().round(3)

    # 合併平均值和方差成 "M (SD)" 形式
    mean_std_A = mean_A.astype(str) + " (" + std_A.astype(str) + ")"
    mean_std_V = mean_V.astype(str) + " (" + std_V.astype(str) + ")"
    mean_std_C = mean_C.astype(str) + " (" + std_C.astype(str) + ")"
    mean_std = mean_std_C + " & " + mean_std_V + " & " + mean_std_A

    # 將結果整理成DataFrame
    results_A_mean_std = pd.DataFrame({'Mean_Std_A': mean_std_A})
    results_V_mean_std = pd.DataFrame({'Mean_Std_V': mean_std_V})
    results_C_mean_std = pd.DataFrame({'Mean_Std_C': mean_std_C})
    results_mean_std = pd.DataFrame({'Mean_Std': mean_std})

    # combined_results = pd.concat([results_A_mean_var, results_N_mean_var], axis=1)
    combined_results = pd.concat([results_C_mean_std, results_V_mean_std, results_A_mean_std, results_mean_std], axis=1)

    display(combined_results)
```
