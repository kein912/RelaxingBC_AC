```python
def add_ratio_columns(df):
    # 检查需要的列是否在DataFrame中
    required_columns = [
        'FC3_theta', 'FC3_beta', 
        'FCz_theta', 'FCz_beta', 
        'Pz_theta', 'Pz_beta', 
        'Oz_theta', 'Oz_beta'
    ]
    missing_columns = [col for col in required_columns if col not in df.columns]
    if missing_columns:
        raise ValueError(f"Missing columns in DataFrame: {missing_columns}")

    # 添加新的列
    # 使用.loc确保操作在原DataFrame上执行
    df.loc[:, 'FC3_theta/beta'] = df['FC3_theta'] / df['FC3_beta']
    df.loc[:, 'FCz_theta/beta'] = df['FCz_theta'] / df['FCz_beta']
    df.loc[:, 'Pz_theta/beta'] = df['Pz_theta'] / df['Pz_beta']
    df.loc[:, 'Oz_theta/beta'] = df['Oz_theta'] / df['Oz_beta']
```

```python
def avgstd(type, path, group):
    # 3group
    mean_data = pd.read_excel("%s/%s_mean.xlsx"%(path, type))
    std_data = pd.read_excel("%s/%s_Std.xlsx"%(path, type))
    # print(mean_data['data_name'] == (group + "_Pre"))

    premean = mean_data[mean_data['data_name'] == (group + "_Pre")].drop('data_name', axis=1).round(2)
    prestd = std_data[std_data['data_name'] == (group + "_Pre")].drop('data_name', axis=1).round(2)
    postmean = mean_data[mean_data['data_name'] == (group + "_Post")].drop('data_name', axis=1).round(2)
    poststd = std_data[std_data['data_name'] == (group + "_Post")].drop('data_name', axis=1).round(2)

    # 合併平均值和方差成 "Mean (Std)" 形式
    mean_std_pre = premean.astype(str) + " (" + prestd.astype(str) + ")"
    mean_std_post = postmean.astype(str) + " (" + poststd.astype(str) + ")"

    # 將結果整理成DataFrame
    results_pre_mean_std = mean_std_pre.T
    results_pre_mean_std.columns = ['Mean_Std_pre']
    results_post_mean_std = mean_std_post.T
    results_post_mean_std.columns = ['Mean_Std_post']

    combined_results = pd.concat([results_pre_mean_std, results_post_mean_std], axis=1)
    combined_results['%s Mean_Std'%group] = combined_results.apply(lambda x: f"{x['Mean_Std_pre']} & {x['Mean_Std_post']} & ", axis=1)

    display(combined_results)
```
